# Cancer detection 

## imports

```python
import copy, math
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from functions import*
from sklearn.linear_model import LogisticRegression
```

### Data and data visvalization

##### Data set

input size of tumour

```python
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])
```

#### data view

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
plot_data(X_train, y_train, ax)

ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)
plt.show()
```

### The Model

```python

lr_model = LogisticRegression()
```

### Training

```python
lr_model.fit(X_train , y_train)
```

## Prediction

```python
y_pred = lr_model.predict(X_train)

print("Prediction on training set:", y_pred)
```

```python
# Extract weights (coefficients) and intercept (bias)
w_out = lr_model.coef_  # Coefficients
b_out = lr_model.intercept_  # Intercept
print(w_out)
print(b_out)
```

##### Decision bountry

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
plot_data(X_train, y_train, ax)

ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)

w1 = w_out[0,0]  # Extracted from w_out[0,0]
w2 = w_out[0,1]  # Extracted from w_out[0,1]
b = b_out  # Bias

# Calculate the decision boundary
x_vals = np.linspace(0, 4, 100)  # Range for x0
decision_boundary = -(w1 * x_vals + b) / w2  # Solve for x1

# Plot the decision boundary
ax.plot(x_vals, decision_boundary, color='blue', lw=1)

# Display the plot
plt.show()
```

# Accuracy

```python
# Calculate the accuracy on the training set
accuracy = lr_model.score(X_train, y_train) * 100

# Print the accuracy as a percentage
print(f"Accuracy on training set: {accuracy:.2f}%")
```
